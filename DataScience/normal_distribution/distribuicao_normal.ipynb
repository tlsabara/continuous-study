# Distribuição normal ou gaussiana

A distribuição normal é a(distribuição) mais importante da estatistica. E sua composição tende ao simetrismo, proximo ou na média.

Quando plotado em um gráfico apresenta uma forma de "sino"

```python
import requests
from PIL import Image
import os
from pathlib import Path
from io import BytesIO
```

```python
ROOT_DIR = Path(os.path.abspath(os.path.curdir)).parent
url_img_normal_distr = "https://t3.ftcdn.net/jpg/04/78/33/96/360_F_478339655_b125oNLiT5oN4QygBeaWfnDVOVH2EfIX.jpg"
```

```python
response = requests.get(url_img_normal_distr, stream=True)
response.raw.decode_content = True
```

```python
img =  Image.open( BytesIO(response.content))
img.save(ROOT_DIR / 'oi.png')
```
